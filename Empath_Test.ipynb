```python
import pandas as pd
from empath import Empath
from pprint import pprint
lexicon = Empath()

def triage_emotions(texts, categories=None):
    results = []
    for text in texts:
        analysis = lexicon.analyze(text, categories=categories, normalize=True)
        results.append(analysis)
    return pd.DataFrame(results)

df = pd.read_csv("balanced_english_data.csv")
texts = df['Description'].tolist()

emo_categories = [
    "joy",  "trust", "anticipation", "surprise"
]
emo_df = triage_emotions(texts, categories=emo_categories)
emo_df["text"] = texts
emo_df["total_emotion"] = emo_df[emo_categories].sum(axis=1)

threshold = 0.1
high_emo_rows = emo_df[emo_df["total_emotion"] > threshold]
pprint(high_emo_rows.iloc[:3, :])
```
